```python
import pandas as pd
import sqlite3
```

1. Add file CSV

```python
users = pd.read_csv('../data/raw/users.csv')
campaigns = pd.read_csv('../data/raw/campaigns.csv')
ads = pd.read_csv('../data/raw/ads.csv')
ad_events = pd.read_csv('../data/raw/ad_events.csv')
```

3. Check for duplicates

```python
print("Users duplicate:", users.duplicated().sum())
print("Campaigns duplicate:", campaigns.duplicated().sum())
print("Ads duplicate:", ads.duplicated().sum())
print("Ad Events duplicate:", ad_events.duplicated().sum())
```

4. Basic data visualization

```python
import matplotlib.pyplot as plt
```

4.1. Distribution of ad event types

```python
ad_events['event_type'].value_counts().plot(kind='bar')
plt.title('Distribution of Ad Event Types')
plt.xlabel('Event Type')
plt.ylabel('Count')
plt.show()
```

4.2. User distribution

```python
users['age_group'].value_counts().plot(kind='bar')
plt.title('User Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Users')
plt.show()
```

4.3. Campaign distribution by platform

```python
ads['ad_platform'].value_counts().plot(kind='bar')
plt.title('Ads Distribution by Platform')
plt.xlabel('Platform')
plt.ylabel('Number of Ads')
plt.show()
```
